```python
# 2. Pass / Fail Analyzer

#marks
marks = [45, 78, 90, 33, 60]

# Counting students who scored 50 or more
pass_count = sum(1 for m in marks if m >= 50)

# Counting students who scored less than 50
fail_count = sum(1 for m in marks if m < 50)

# Printing results
print("Passed:", pass_count)
print("Failed:", fail_count)
```

```python
# 3. Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

# Removing extra spaces and converting to lowercase
cleaned = [name.strip().lower() for name in names]

# Printing cleaned names
print("Cleaned Names:", cleaned)
```

```python
# 4. Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

# Checking each message/Finding length of message
for msg in messages:
    print("Message:", msg)
    print("Length:", len(msg))   
    
# Checking if message is longer than 10 characters
    if len(msg) > 10:
        print("Long message")
print()
```

```python
# 5. Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Counting how many times "ERROR" appears
error_count = logs.count("ERROR")

# Printing total error count
print("Total ERROR:", error_count)
```
