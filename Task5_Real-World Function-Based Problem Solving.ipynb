**Problem 1: Social Media – Post Engagement Analyzer**


```python
def post_engagement_analyzer(likes_list):
    total_likes = 0
    
    # Loop to calculate total likes
    for like in likes_list:
        total_likes += like
    
    # Conditional logic
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    # Output
    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Example Input
likes = [200, 300, 150, 400]

post_engagement_analyzer(likes)
```

**Problem 3: Agriculture – Rainfall Adequacy Checker**


```python
def rainfall_checker(rainfall_list, required_level):
    
    total = 0   # variable to store total rainfall
    
    # Loop to calculate total rainfall
    for rain in rainfall_list:
        total = total + rain
    
    # Calculate average rainfall
    average = total / len(rainfall_list)
    
    # Check if average meets required level
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    # Display results
    print("Average Rainfall:", average)
    print("Rainfall Status:", status)


# Example seasonal rainfall data
rainfall_data = [70, 75, 80, 65, 70]

# Required rainfall level
required = 70

# Call the function
rainfall_checker(rainfall_data, required)
```

**Problem 4: Social Media – Duplicate Account Detection**


```python
def detect_duplicates(usernames):
    
    # Convert list to set (removes duplicates automatically)
    unique_usernames = set(usernames)
    
    # Compare lengths
    if len(usernames) != len(unique_usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example data
users = ["rahul123", "priya_01", "rahul123", "amit99"]

# Call function
detect_duplicates(users)
```

**Problem 6: Agriculture – Premium Crop Price Filter**

```python
def premium_crop_filter(prices):
    # empty list to store premium prices
    premium_crops = []  
    
    # Loop through each crop price
    for price in prices:
        
        # Check if price is greater than 2000
        if price > 2000:
            premium_crops.append(price)   # add to premium list
    
    # Display premium crops
    print("Premium Crops:", premium_crops)


# Example data
crop_prices = [1500, 2500, 1800, 3200]

# Call function
premium_crop_filter(crop_prices)
```

**Problem 7: System Monitoring – Application Health Checker**


```python
def application_health_checker(error_count):
    
    # Check system health based on error count
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    # Display results
    print("Error Count:", error_count)
    print("System Status:", status)


# Example
application_health_checker(7)
```

**Problem 8: Banking – Daily Transaction Limit Checker**


```python
def check_transaction(amount):
    
    daily_limit = 50000   # set daily transaction limit
    
    # Check whether transaction is within limit
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"
    
    # Display result
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)


# Call the function to get output
check_transaction(60000)
```

**Problem 9: E-Learning – Student Attendance Eligibility System**


```python
def check_exam_eligibility(attendance_list):
    # total number of classes
    total_classes = len(attendance_list)
    # counter for attended classes
    attended = 0  
    
    # Loop to count attended classes
    for status in attendance_list:
        if status == "P":      
            attended += 1
    
    # Calculate attendance percentage
    percentage = (attended / total_classes) * 100
    
    # Check eligibility
    if percentage >= 75:
        result = "Eligible"
    else:
        result = "Not Eligible"
    
    # Display result
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", result)


# Example attendance record (P = Present, A = Absent)
attendance_data = ["P", "P", "A", "P", "P"]

check_exam_eligibility(attendance_data)
```

**Problem 10: Smart Electricity Bill Analyzer**


```python
def calculate_electricity_bill(units):
    
    # Calculate bill using slab system
    if units <= 100:
        bill = units * 3
    
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    
    # Classify usage based on total bill
    if bill < 500:
        status = "Low Usage"
    
    elif bill <= 1500:
        status = "Moderate Usage"
    
    else:
        status = "High Usage"
    
    
    return bill, status


# Function call
bill_amount, usage_status = calculate_electricity_bill(250)

print("Total Bill:", bill_amount)
print("Usage Level:", usage_status)
```
