**Problem 1: Mobile Recharge Validation System**

```python
# List of valid recharge plans
valid_plans = [199, 299, 399, 599]

# Infinite loop to keep asking until valid input is given
while True:

    # Take recharge amount input from user
    amount = int(input("Enter recharge amount: "))
    
    # Take plan selection input from user
    plan = int(input("Enter plan (199, 299, 399, 599): "))

    # Check if entered amount is below minimum requirement
    if amount < 50:
        print("Amount must be at least ₹50")

    # Check if selected plan is not in the valid plans list
    elif plan not in valid_plans:
        print("Invalid plan selected")

    # If both conditions are satisfied
    else:
        print("Recharge Successful!")
        break   # Exit loop after successful recharge

    # Executes if recharge was not successful
    print("Please try again.\n")
```

**Problem 2: Inventory Reorder Alert System**


```python
# Dictionary of products and stock
products = {"Rice": 10,"Sugar": 25,"Oil": 5,"Wheat": 20}
# Function to check inventory stock
def check_inventory(products):
    # Loop through dictionary
    for product, stock in products.items():

        # Check if stock is less than 15
        if stock < 15:
            print(product, "- Reorder Alert")
        else:
            print(product, "- Stock OK")
# Call the function
check_inventory(products)
```

**Problem 3: Student Result Processing System**


```python
# Function to process student results
def process_result(marks):
    
    # Variable to store total marks
    total = 0
    
    # Initialize index for while loop
    i = 0

    # Loop through the list using index
    while i < len(marks):
        total += marks[i]   # Add each mark to total
        i += 1              # Move to next index

    # Calculate average marks
    average = total / len(marks)

    # Check if student passed or failed
    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# Call the function with example marks
print(process_result([60, 45, 70, 55]))
```

**Problem 4: Cab Fare Estimator with Retry Option**


```python
# Function to calculate cab fare
def calculate_fare(distance, peak):

    base_fare = 50              # Fixed base fare
    per_km_rate = 12            # Cost per kilometer

    fare = base_fare + (distance * per_km_rate)

    # Add 25% extra if peak hour
    if peak == "yes":
        fare = fare + (fare * 0.25)

    return fare


# While loop for retry option
while True:

    # Take user input
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ")

    # Call function
    total_fare = calculate_fare(distance, peak)

    print("Total Fare: ₹", total_fare)

    # Ask user if they want to calculate again
    retry = input("Do you want to calculate again? (yes/no): ")

    if retry != "yes":
        break
```

**Problem 5: Employee Attendance Eligibility Checker**


```python
# Function to check attendance eligibility
def check_eligibility(attendance):

    present = 0   # Variable to count present days

    # Loop to count "P"
    for day in attendance:
        if day == "P":
            present += 1

    # Calculate attendance percentage
    percentage = (present / len(attendance)) * 100

    # Check eligibility
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Example attendance list
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]

# Call function and print result
result = check_eligibility(attendance)
print("Attendance Status:", result)
```

**Problem 6: Password Strength Checker**


```python
# Function to check password strength
def check_password(password):

    has_digit = False        # To check if digit exists
    has_special = False      # To check if special character exists

    # Loop through each character in password
    for char in password:

        # Check if character is a digit
        if char.isdigit():
            has_digit = True

        # Check if character is one of the special characters
        if char in "@#$":
            has_special = True

    # Final condition check
    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


# Take input from user
password = input("Enter password: ")

# Call function and print result
print(check_password(password))
```
