**Problem Statement 1: Employee Performance Bonus Eligibility**


```python
# Employee performance scores
employees = { "Ravi": 92, "Anita": 88, "Kiran": 92, "Suresh": 85}
# Find highest score
highest_score = max(employees.values())
# Check who has the highest score
for name in employees:
    if employees[name] == highest_score:
        print(name, "is eligible for bonus")
# Print highest score
print("Score:", highest_score)
```

**Problem Statement 2: Search Query Keyword Analysis**


```python
# Given search query
query = "Buy mobile phone buy phone online"
#Convert to lowercase
query = query.lower()
#Remove punctuation (not needed here, but added for safety)
query = query.replace(".", "").replace(",", "")
#Split sentence into words
words = query.split()
#Count frequency of each word
word_count = {}
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1
#Display words searched more than once
result = {}
for word in word_count:
    if word_count[word] > 1:
        result[word] = word_count[word]
print(result)
```

**Problem Statement 3: Sensor Data Validation**

```python

# List of sensor readings (index = hour, value = reading)
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
# List to store valid readings
valid_readings = []
# Loop through the list using index
for i in range(len(sensor_readings)):
# Check if reading is even
    if sensor_readings[i] % 2 == 0:
# Store as (hour_index, reading_value)
       valid_readings.append((i, sensor_readings[i]))
# Display result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

**Problem Statement 4: Email Domain Usage Analysis**


```python
# List of employee emails
emails = [ "ravi@gmail.com", "anita@yahoo.com", "kiran@gmail.com", "suresh@gmail.com", "meena@yahoo.com"]
# Dictionary to store domain count
domain_count = {}
# Total number of emails
total = len(emails)
#Extract domain and count
for email in emails:
    # Split at '@' and take domain part
    domain = email.split("@")[1]
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1
#Calculate and display percentage
for domain in domain_count:
    percentage = (domain_count[domain] / total) * 100
    print(domain + ":", str(int(percentage)) + "%")
```

**Problem Statement 5: Sales Spike Detection**


```python
# List of daily sales
sales = [1200, 1500, 900, 2200, 1400, 3000]
#Calculate average sales
average =  sum(sales)/ len(sales)
#Find 30% above average
limit = average * 1.25
# Check which days exceed the limit
for i in range(len(sales)):
    if sales[i] > limit:
        print("Day", i + 1, ":", sales[i])
```

**Problem Statement 6: Duplicate User ID Detection**

```python
# List of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
# Dictionary to store count of each user ID
count = {}
#Count occurrences
for user in user_ids:
    if user in count:
        count[user] += 1
    else:
        count[user] = 1
#Display only duplicates
for user in count:
    if count[user] > 1:
        print(user, "→", count[user], "times")
```
